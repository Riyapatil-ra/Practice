```python
nums=[1,2,2,3,4,4,5,6,5,7,8,1,4]
unique=[]
count=0
for i in nums:
    if i not in unique:
        unique.append(i)
    else:
        count+=1
print("Unique elements list:",unique)
print("Total count of duplicates elements: ",count)
```
