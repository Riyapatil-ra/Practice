```python
b1=np.arange(1,11)
b2=np.arange(12,22)

sum=np.add(b1,b2)
mul=np.multiply(b1,b2)
diff=np.subtract(b1,b2)
div=np.divide(b1,b2)

print(f"b1 is {b1}")
print(f"b2 is {b2}")
print(f"Sum of b1 and b2 is {sum}")
print(f"Multiplication of b1 and b2 is {mul}")
print(f"Difference of b1 and b2 is {diff}")
print(f"Division of b1 and b2 is {div}")
```

```python
c1=np.arange(22,40)

print(c1)
print(f"Summation of c1 is {np.sum(c1)}")
print(f"Mean of c1 is {np.mean(c1)}")
print(f"Minimum of c1 is {np.min(c1)}")
print(f"Maximum of c1 is {np.max(c1)}")
```

```python
c3=np.random.randint(0,100,9).reshape(3,3)
print(c3)

print("Only the first 2 rows and 2 columns:\n", c3[:2,:2])
```
